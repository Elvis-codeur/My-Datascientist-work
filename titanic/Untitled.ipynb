```python
columns_to_use = ["sex","class"]
correspandance = {}
correspandance["sex"] = {"male":0,"female":10}
correspandance["class"] = {"Third":0,"Second":5,"First":10}
```

```python
def prepare_dataframe(df):
    result = {}
    for c in columns_to_use:
        cor = correspandance[c] # Les correspondances
        col = df[c] # Les colonnes
        for u in cor:
            col = col.replace(u,cor[u])
            
        result[c] = col
        
    return pd.DataFrame.from_dict(result)
        
            
```

```python
final_train_data = prepare_dataframe(dftrain)
y_train = dftrain["survived"].values
```

```python
final_test_data = prepare_dataframe(dfeval)
y_test = dfeval["survived"].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg =LinearRegression().fit(final_train_data.values,y_train)
```

```python
reg.score(final_test_data,y_test)
```
